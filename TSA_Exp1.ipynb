```python
import pandas as pd
import numpy as np
```

```python
from statsmodels.tsa.filters.hp_filter import hpfilter
```

```python
from pandas import datetime
from matplotlib import pyplot
```

**India Exchange Rate dataset**

```python
df1 = pd.read_excel('/content/India_Exchange_Rate_Dataset.xls', index_col=0,parse_dates=True)
```

lambda=100

```python
EXINUS_cycle,EXINUS_trend = hpfilter(df1['EXINUS'], lamb=100)
EXINUS_trend.plot(figsize=(15,6)).autoscale(axis='x',tight=True)
```

```python
EXINUS_cycle.plot(figsize=(15,6)).autoscale(axis='x',tight=True)
```

lambda=1600

```python
EXINUS_cycle,EXINUS_trend = hpfilter(df1['EXINUS'], lamb=1600)
EXINUS_trend.plot(figsize=(15,6)).autoscale(axis='x',tight=True)
```

lambda=14400

```python
EXINUS_cycle,EXINUS_trend = hpfilter(df1['EXINUS'], lamb=14400)
EXINUS_trend.plot(figsize=(15,6)).autoscale(axis='x',tight=True)
```

Default Differencing from Scratch

```python
X = df1.values
diff = list()
for i in range(1, len(X)):
  value = X[i] - X[i - 1]
  diff.append(value)
pyplot.figure(figsize=(15,6))
pyplot.plot(diff)
pyplot.title('Detrending from scratch', fontsize=16)
pyplot.xlabel('Year')
pyplot.ylabel('EXINUS exchange rate')
pyplot.show()
```

Default Differencing using library

```python
pyplot.figure(figsize=(15,6))
pyplot.plot(df1.diff())
pyplot.title('Detrending using Pandas Differencing', fontsize=16)
pyplot.xlabel('Year')
pyplot.ylabel('EXINUS exchange rate')
pyplot.show()
```

**Detrending using Scipy Signal**

```python
from scipy import signal
import warnings
```

```python
warnings.filterwarnings("ignore")
detrended = signal.detrend(df1.EXINUS.values)
pyplot.figure(figsize=(15,6))
pyplot.plot(detrended)
pyplot.xlabel('EXINUS')
pyplot.ylabel('Frequency')
pyplot.title('Detrending using Scipy Signal', fontsize=16)
```

**Detrending using HP filter**

```python
import warnings
warnings.filterwarnings("ignore")
df1['trend'] = EXINUS_trend
detrended = df1.EXINUS - df1['trend']
pyplot.figure(figsize=(15,6))
pyplot.plot(detrended)
pyplot.title('Detrending using HP Filter', fontsize=16)
pyplot.xlabel('Year')
pyplot.ylabel('EXINUS exchange rate')
pyplot.show()
```
